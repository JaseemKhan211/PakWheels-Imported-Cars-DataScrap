# Finalize, iterate over all links and get complete data

```python
for x in range(1, 5):
        if x == 1:
            # The webpage URL
            url = 'https://www.pakwheels.com/used-cars/imported/57428?'
        else:
            url = 'https://www.pakwheels.com/used-cars/imported/57428?page=' + str(x)

        # HTTP Request
        webpage = requests.get(url, headers = header)

        # Soup object containing all data
        soup = BeautifulSoup(webpage.content, "html.parser")

        links_list = []
        articles_info = {"vihicleName":[], "modelYear":[], "millage":[], "fuelType":[], "transmission":[]}
        links = soup.find_all("a", attrs = {'class' : 'car-name ad-detail-path'})

        # Loops for extracting links from Tag Objects
        for link in links:
            links_list.append(link.get('href'))

        #loops for child pages
        for link in links_list:
            new_webpage = requests.get('https://www.pakwheels.com/' + link, headers = header)
            new_soup = BeautifulSoup(new_webpage.content, 'html.parser')

            title = new_soup.find("h1").text
            articlesDetail = new_soup.find("table", attrs = {"class":'table table-bordered text-center table-engine-detail fs16'}).find_all('p')

            year = articlesDetail[0].text.strip()
            millage = articleDetail[1].text.strip()
            fuelType = articleDetail[2].text.strip()
            transmission = articleDetail[3].text.strip()

            articles_info['vihicleName'].append(title)
            articles_info['modelYear'].append(year)
            articles_info['millage'].append(millage)
            articles_info['fuelType'].append(fuelType)
            articles_info['transmission'].append(transmission)

vehiclesDf = pd.DataFrame.from_dict(articles_info)
vehiclesDf.to_csv("SCRAPPING/pakWheels_importCars.csv", header = True, index = False)
```
